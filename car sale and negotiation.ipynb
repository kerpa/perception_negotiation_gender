```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

### Definition of the common steps in car sale process
I regroup different website where is mentioned the steps needed to buy a car, this website are insiders

```python
path = 'websitedatacars.csv'
dfwebsite = pd.read_csv(path)
dfwebsite
```

```python
dfwebsite.describe()
```

```python
List_steps=["Research info","Prepare budget","Compare offers","Drive the car","Negotiate the deal","Delivery of the car"]
```

### Importance of each steps for insiders

```python
List_proportions1=[]
List_proportions1.append(dfwebsite["Research"].mean())
List_proportions1.append(dfwebsite["Prepare Budget"].mean())
List_proportions1.append(dfwebsite["Compare offers"].mean())
List_proportions1.append(dfwebsite["Drive the car"].mean())
List_proportions1.append(dfwebsite["Negotiate the deal"].mean())
List_proportions1.append(dfwebsite["Take on delivery"].mean())
dfdatawebsite = pd.DataFrame(list(zip(List_steps, List_proportions1)), columns=["Step", "percentage"])
dfdatawebsite
```

```python
dfwebsite.corr()
```

### Analysis of the questionnaire

```python
path2 = 'data_car_survey_raw.csv'
dfdata = pd.read_csv(path2)
dfdata.columns = ["car","driving license","Research info","Prepare budget","Compare offers","Drive the car","Negotiation","Delivery","gender","age","country"] 
dfdata.head()
```

```python
dfdata.describe()
```

```python
fig4 = dfdata['gender'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
fig5 = dfdata['age'].value_counts().plot(kind='pie',autopct='%1.1f%%')
```

```python
correlation_matrix = dfdata.corr()
correlation_matrix
```

```python
sns.heatmap(data = correlation_matrix, annot=True)
```

```python
List_proportions2=[]
List_proportions2.append(dfdata["Research info"].mean()/5)
List_proportions2.append(dfdata["Prepare budget"].mean()/5)
List_proportions2.append(dfdata["Compare offers"].mean()/5)
List_proportions2.append(dfdata["Drive the car"].mean()/5)
List_proportions2.append(dfdata["Negotiation"].mean()/5)
List_proportions2.append(dfdata["Delivery"].mean()/5)
```

```python
dfsteps = pd.DataFrame(list(zip(List_steps, List_proportions2)), columns=["Step", "percentage"])
dfsteps
```

```python
fig2 = sns.barplot(x=List_proportions2,y=List_steps)
plt.xlabel('Rating')
plt.ylabel('Step')
```

```python
dfdata.groupby(['gender']).mean()
```

```python
f1 = dfdata.groupby(['gender']).mean().plot(kind='bar',figsize=(12,6),ylabel='Rating')
f1.grid()
plt.xticks(rotation=45)
```
